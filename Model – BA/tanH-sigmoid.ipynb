```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## create plot of TanH function

# define the function
def tanh(x):
    return np.tanh(x)

def sigmoid(x):
    return 1/(1+np.exp(-x))

## Create two subplots for the two functions
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
plt.subplots_adjust(wspace=0.5)
fig.suptitle('Activation Functions', fontsize=16)


## Plot the Sigmoid function add gridlines
x = np.linspace(-10, 10, 100)
ax1.plot(x, sigmoid(x), label='sigmoid')
ax1.set_title('A: Sigmoid')
ax1.grid()
ax1.set_xlabel('x')
ax1.set_ylabel('y')


## Plot the TanH function add gridlines
x = np.linspace(-10, 10, 100)
ax2.plot(x, tanh(x), label='tanh', color = 'orange')
ax2.set_title('B: Tanh')
ax2.grid()
ax2.set_xlabel('x')
ax2.set_ylabel('y')

plt.savefig('tanh-sigmoid.png')

plt.show()





```
